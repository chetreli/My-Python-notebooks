```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
crash_df  = pd.read_csv('car_crashes.csv')
crash_df
```

Distribution plot

```python
sns.distplot(crash_df['not_distracted'])
```

Distribution plot without KDE

```python
sns.distplot(crash_df['not_distracted'], kde = False)
```

```python
sns.distplot(crash_df['not_distracted'], kde = False, bins = 10)
```

Joint plot

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex')
```

KDE plot

```python
sns.kdeplot(crash_df['alcohol'])
```

 Pair plots

```python
sns.pairplot(crash_df)
```
